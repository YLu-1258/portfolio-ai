```python
import yfinance as yf
import pandas as pd
import pickle
from datetime import datetime, timedelta

# Define the ticker symbol
ticker_symbol = "DUOL"

# Create a Ticker object
ticker = yf.Ticker(ticker_symbol)

# https://ranaroussi.github.io/yfinance/reference/api/yfinance.download.html#yfinance.download
# df = pd.DataFrame(ticker.history(period="max", interval="1m"))
iterations = 0
end = datetime.now()
histories = []
while iterations < 7:
    history = ticker.history(start = end - timedelta(days=4), end = end, interval="1m")
    histories.append(history)
    end = end - timedelta(days=4)
    iterations += 1

data = pd.concat(histories)
data
# data.sort_values("Datetime")
# data
```

```python
data = data.sort_index()
```

```python
with open('DUOL.pkl', 'wb') as file:
    pickle.dump(data, file)
```
